## Aggregation
Performing the same op on a number of things

```python
rand = np.random.random(10000)
%timeit np.sum(rand) # NPs sum
%timeit sum(rand) # Py sum
```

```python
# STD - how spread out a group of nos is from mean
np.std(rand)
```

```python
#Variance - measure of avg deg of difference between each no and mean in a data set
np.var(rand)
```

```python
# sd = sqrt of var
np.sqrt(np.var(rand))
```

```python
# DEMO SD and VARIANCE
high_var = np.array([1,100,200,1000,8000])
low_var = np.array([2,4,6,8,10])
np.var(high_var), np.var(low_var)
```

## Dot Product

```python
np.random.seed(10)
mat1 = np.random.randint(10, size=(5,3))
mat2 = np.random.randint(10, size=(5,3))
mat1, mat2
```

```python
mat1 = mat1.T
np.dot(mat1, mat2)
```

```python
# Dot prod example
np.random.seed(10)
sales_amt =  np.random.randint(20, size=(5,3))
sales_amt
```

```python
weekly_sales = pd.DataFrame(sales_amt, index=["Mon", "Tue", "Wed", "Thu", "Fri"], columns=["Almond","Peanut","Cashew"])
weekly_sales
```

```python
prices = np.array([10, 8, 12])
prices.shape
```

```python
butter_prices = pd.DataFrame(prices.reshape(1,3), index=["Price"], columns=["Almond","Peanut","Cashew"])
butter_prices
```

```python
total_sales = prices.dot(sales_amt.T)
total_sales
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
